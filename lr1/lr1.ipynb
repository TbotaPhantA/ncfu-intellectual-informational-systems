```python
import numpy as np
# Загрузка файла
switzerlandData = np.genfromtxt("./heartDisease/processed.switzerland.data", delimiter=",")

# Возраст средний, минимальный, максимальный
age_avg = np.mean(switzerlandData[:, 0])
age_min = np.min(switzerlandData[:, 0])
age_max = np.max(switzerlandData[:, 0])
print("Age:")
print("  Average:", age_avg)
print("  Minimum:", age_min)
print("  Maximum:", age_max)

# посчитать кол-во типов грудных болей
typical_angina_count = np.count_nonzero(switzerlandData[:, 2] == 1)
atypical_angina_count = np.count_nonzero(switzerlandData[:, 2] == 2)
non_anginal_pain_count = np.count_nonzero(switzerlandData[:, 2] == 3)
asymptomatic_count = np.count_nonzero(switzerlandData[:, 2] == 4)

total_count = len(switzerlandData)

# вычисления пропорций типов болей
typical_angina_percentage = (typical_angina_count / total_count) * 100
atypical_angina_percentage = (atypical_angina_count / total_count) * 100
non_anginal_pain_percentage = (non_anginal_pain_count / total_count) * 100
asymptomatic_percentage = (asymptomatic_count / total_count) * 100

print("Chest Pain Type Percentages:")
print("  Typical Angina:", typical_angina_percentage)
print("  Atypical Angina:", atypical_angina_percentage)
print("  Non-Anginal Pain:", non_anginal_pain_percentage)
print("  Asymptomatic:", asymptomatic_percentage)
```
