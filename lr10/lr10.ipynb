```python
# импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Загрузка данных c ценами страховки
data = pd.read_csv('./ice_cream.csv')
data.head()
```

```python
# визуализация
data.describe()
```

```python
# Обработка пропущенных значений
from sklearn.impute import SimpleImputer

data.sort_values(by='Temperature (°C)', ascending=True)

# индекс веса тела
X = data.loc[:,'Temperature (°C)':'Temperature (°C)']
# стоимость
y = data['Ice Cream Sales (units)']

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

print(X_imputed)
```

```python
# Обучение линейной модели
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Обучение полиномиальной модели
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=10)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Предсказание, обработка и визуализация результатов
y_pred_lin = lin_reg.predict([[6.5]])
y_pred_poly = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print(y_pred_lin, y_pred_poly)
# Визуализация линейном модели
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Ice cream sales')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

- На данном графике видно, что для таких данных одной линией трудно подстроится под данные

```python
# Визуализация полиномиальной модели
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Ice cream sales')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

- На данном графике видно, что если воспользоваться полиномеальной моделью, то график может подстроится под более сложный и нелинейный набор данных

```python
# Обучение полиномиальной модели при степени n = 6
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=6)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

# Визуализация
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Ice cream sales')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```
